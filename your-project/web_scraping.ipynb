```python
from bs4 import BeautifulSoup as soup
import requests
import pandas as pd
```

```python
url = 'http://www.worldstopexports.com/wine-exports-country/'
```

```python
response = requests.get(url)
soup = soup(response.content, "html.parser")
soup
```

```python
num_int = len(soup.find_all("td", class_ = "column-2"))

exporter = []

for i in range(num_int):
    countries = soup.find_all("td", class_ = "column-2")[i].get_text()
    exporter.append(countries)
    
exporter    
```

```python
num_int = len(soup.find_all("td", class_ = "column-3"))

exported_wine = []

for i in range(num_int):
    q = soup.find_all("td", class_ = "column-3")[i].get_text()
    exported_wine.append(q)
    
exported_wine  
```

```python
num_int = len(soup.find_all("td", class_ = "column-4"))

variation = []

for i in range(num_int):
    z = soup.find_all("td", class_ = "column-4")[i].get_text()
    variation.append(z)
    
variation 
```

```python
wine_exports = pd.DataFrame(list(zip(exporter, exported_wine, variation)), 
               columns = ['Exporter', 'Exported_wine(US$)', 'Variation']) 
wine_exports
```
